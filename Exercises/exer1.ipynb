## Αφαίρεση σημείων στίξης

```python
import string
string.punctuation # εμφανίζει τα σημεία στίξης
```

```python
# Αφαίρεση σημείων στίξης απο τα tokens
sentence = "Monticello wasn't designated as UNESCO World Heritage Siteuntil 1987."
tokens = nltk.word_tokenize(sentence)
cleaned_tokens=[]
for token in tokens:
    if token not in string.punctuation:
        cleaned_tokens.append(token)
cleaned_tokens
```

## Αφαίρεση προθεμάτων

```python
# Για να δείτε τα προθήματα της Αγγλικής γλώσσας
stopwords = nltk.corpus.stopwords.words('english')
stopwords
```

```python
# Αφαίρεση προθημάτων απο τα tokens
cleaned_tokens=[]
for token in tokens:
    if token not in stopwords:
        cleaned_tokens.append(token)
cleaned_tokens
```

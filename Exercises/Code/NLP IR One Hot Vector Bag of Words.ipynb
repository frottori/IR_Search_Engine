# Ας φτιαξουμε πίνακα από μία πρόταση (NLPIA pp44 (38))

```python
sentence = "Thomas Jefferson began building Monticello at the age of 26."
```

```python
import numpy as np
token_sequence=str.split(sentence)
#token_sequence
#Αν θέλω ολόκληρο κείμενο...
#token_sequence=text1.tokens
vocab = sorted(set(token_sequence)) #φτιάξε το λεξιλόγιο αφαιρώντας τα tokens που επαναλαμβάνονται
', '.join(vocab) #Ενώνει τις λέξεις του vocab με κόμα απλά για να τις δούμε
```

```python
num_tokens = len(token_sequence)
vocab_size = len(vocab)
#φτιάξε με μηδενικά ένα vector με διαστάσεις αρ_tokens X μέγεθος_λεξιλογίου
onehot_vectors = np.zeros((num_tokens, vocab_size), int)

onehot_vectors
```

```python
for i, word in enumerate(token_sequence):
    onehot_vectors[i, vocab.index(word)] = 1
' '.join(vocab)
```

```python
onehot_vectors
```

Εμφάνισέ τα πιο όμορφα με το pandas 

```python
#pip install pandas

import pandas as pd
pd.DataFrame(onehot_vectors, columns=vocab)
```

```python
#NLPIA section 2.2 
sentence_bow = {} #create a dictionary
sentence = "Thomas Jefferson began building Monticello at the age of 26 and at the"
for token in sentence.split():
    wordfreq=sentence.count(token) #μέτρα την συχνότητα που εμφανίζεται το token.......................................
    sentence_bow[token]=wordfreq
sorted(sentence_bow.items())

```

# Χρησιμοποιούμε τρεις προτάσεις ακόμα για να κάνουμε συγκρίσεις
A DataFrame indexes both the columns (documents) and rows (words) so it can be an "inverse index" for document retrieval, in case you want to find a Trivial Pursuit answer in a hurry.

```python
sentences = "Construction was done mostly by local masons and carpenters.\n" \
    "He moved in to the South Pavilion in 1770.\n" \
     "Turning Monticello in to a neoclassical masterpiece was Jefferson's obsession."
corpus = {}
corpus['sent0'] = dict((tok.strip('.'), sentence.count(tok)) for tok in sentence.split())

for i, sent in enumerate(sentences.split('\n')):
    corpus['sent{}'.format(i + 1)] = dict((tok.strip('.'), sentence.count(tok)) for tok in sent.split())
    
df = pd.DataFrame.from_records(corpus).fillna(0).astype(int).T
df[df.columns[:25]] # show just the first 7 tokens (columns)
```

Από τον παραπάνω πίνακα μποροόυμε να δούμε τις συμπτώσεις.
Παρακάτω, μπορούμε να υπολογίσουμε "πρόχειρα" την ομοιότητα δύο εγγράφων με το εσωτερικό τους γινόμενο.

```python
df = df.T
df.sent0.dot(df.sent1)
```

# Αυτό ήταν το πρώτο μας Διανυσματικό μοντέλο  vector space model
(This is your first vector space model (VSM) of natural language documents (sentences).
Not only are dot products possible, but other vector operations are defined for these
bag-of-word vectors: addition, subtraction, OR, AND, and so on. You can even compute
things such as Euclidean distance or the angle between these vectors. This representation
of a document as a binary vector has a lot of power.
